```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import collections
```

### Lag BI, CI, HI & Weather Data

```python
path_dt = './data/full_house_again.csv'
df = pd.read_csv(path_dt)
```

```python
df.head()
```

```python
X = ['Temperature', 'T Max', 'T Min', 'RH', 'Precp']
Y = ['BI', 'CI', 'HI']
YX = ['County_EN', 'YYYY_WW', 'BI', 'CI', 'HI', 'Temperature', 'T Max', 'T Min', 'RH', 'Precp']
```

```python
yx = df[YX]
yx.head()
```

```python
# split by cty
cty_yx = collections.defaultdict(list)

for i in range(yx.shape[0]):

    
    row = yx.iloc[i, :].tolist()
    
    cnty = row[0]
    
    cty_yx[cnty].append(row)
    
```

```python
# convert list to df
cty_yx_df = collections.defaultdict()

for k, v in cty_yx.items():

    x = pd.DataFrame(cty_yx[k], columns = YX)

    cty_yx_df[k] = x
```

```python
cty_yx_df['Taipei'].head()
```

```python
def lag_df(df, lag = 1, y_cols = ['County_EN', 'YYYY_WW', 'BI', 'CI', 'HI'], x_cols = ['Temperature', 'T Max', 'T Min', 'RH', 'Precp']):
    
    # split y, x
    
    YY = df[y_cols]
    XX = df[x_cols]

    n_rows = YY.shape[0]
    
    # slice with df.iloc[:, :]
    
    yy = YY.iloc[lag:, :]
    yy = yy.reset_index()
    xx = XX.iloc[:-lag, :]
    xx = xx.reset_index()
    
    yyxx = pd.concat([yy, xx], axis = 1)
    
    return yyxx

```

```python
lags = 5 # lag 1, 2, 3, 4

for lag in range(1, lags, 1):
    
    print('lag = ', lag)
    
    tmp = []

    for cty in cty_yx_df.keys():
        
        df = cty_yx_df[cty]
        
        df2 = lag_df(df, lag = lag)
        
        tmp.append(df2)
        
    fn = pd.concat(tmp, axis = 0, ignore_index = True)

    fn.to_csv('./data/lag_%s_week_bi_temp.csv' %str(lag))
```

### Lag Dengue Fever & Pop Data

```python
path_dt = './data/denguefever_bi_pop.csv'
df = pd.read_csv(path_dt)
df.head()
```

```python
YX = ['ID', 'Time', 'Denv', 'BI', 'CI', 'HI', 'Pop', 'PopDen']
```

```python
df['ID'] = df.ID.astype(str)
df.dtypes
```

```python
# split by cty
cty_yx = collections.defaultdict(list)

for i in range(df.shape[0]):

    
    row = df.iloc[i, :].tolist()
    
    cnty = row[0]
    
    cty_yx[cnty].append(row)
    
# convert list to df
cty_yx_df = collections.defaultdict()

for k, v in cty_yx.items():

    #print(type(k))
    x = pd.DataFrame(cty_yx[k], columns = YX)

    cty_yx_df[str(k)] = x
```

```python
cty_yx_df['1'].head()
```

```python
lags = 5 # lag 1, 2, 3, 4

y_cols = ['ID', 'Time', 'Denv']
x_cols = ['BI', 'CI', 'HI', 'Pop', 'PopDen']

for lag in range(1, lags, 1):
    
    print('lag = ', lag)
    
    tmp = []

    for cty in cty_yx_df.keys():
        
        df = cty_yx_df[cty]
        
        df2 = lag_df(df, lag = lag, y_cols = y_cols, x_cols = x_cols)
        
        tmp.append(df2)
        
    fn = pd.concat(tmp, axis = 0, ignore_index = True)
    fn = fn[YX]
    fn.to_csv('./data/lag_%s_week_denv_bi_pop.csv' %str(lag))
```
