```python
from tqdm import tqdm
```

```python
# split data for 20 county

cnty_data = collections.defaultdict(list)

for i in tqdm(range(data_distill.shape[0])):
    row = data_distill.iloc[i, :].tolist()
    cnty = row[1]
    if cnty in county:
        #print(row)
        cnty_data[cnty].append(row)
```

```python
# convert data each county from list to dataframe

data_fin = collections.defaultdict()
for k, v in tqdm(cnty_data.items()):
    x = pd.DataFrame(cnty_data[k], columns = use_cols, dtype = float)
    data_fin[k] = x
```

```python
data_fin['高雄市'].tail()
```

```python
# remove 'None' & groupby 'Date'
data_fin2 = {}

for k in tqdm(data_fin.keys()):
    
    tm = data_fin[k].replace(to_replace='None', value=0).dropna()
    
    tm['CI'] = tm.CI.astype(float)
    tm['BI'] = tm.BI.astype(float)
    tm['HI'] = tm.HI.astype(float)
    
    tm['Year'] = pd.to_datetime(tm['Date']).dt.year 

    tm = tm[tm['Year'] >= 2012]

    tm = tm.groupby('Date', as_index=False).agg(['mean'])
    tm.reset_index(level=tm.index.names, inplace=True)
    
    #tm['Year-Week'] = pd.to_datetime(tm['Date']).dt.strftime('%Y-%U')
    
    #tm = tm.groupby('Year-Week', as_index=False).agg(['mean'])
    
    #print(tm2)
    #tm['Year-Week'] = tm['Date'].dt.strftime('%Y-%U')
    
    #tm.reset_index()

    data_fin2[k] = tm
    
```
